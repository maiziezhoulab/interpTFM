```python
from pathlib import Path
import scanpy as sc
import numpy as np

shards=60
p = Path('/maiziezhou_lab2/yunfei/Projects/FM_temp/InterPLM/interplm/ge_shards/cosmx_human_lung_sec8.h5ad')
if not p.exists():
    adata_path = Path("/maiziezhou_lab2/yunfei/Projects/FM_temp/datasets/cosmx/lung/cosmx_human_lung.h5ad")

    adata = sc.read_h5ad(adata_path)

    adata = adata[adata.obs['library_key'] == 7].copy()

    # If adata.X is a sparse matrix (common in Scanpy workflows)
    X = adata.X.toarray() if not isinstance(adata.X, np.ndarray) else adata.X

    # Identify rows where all elements are zero
    all_zero_rows = np.all(X == 0, axis=1)

    # Get indices of all-zero rows
    zero_row_indices = np.where(all_zero_rows)[0]

    # Print number of zero rows and their indices (if any)
    print(f"Number of all-zero rows: {len(zero_row_indices)}")
    if len(zero_row_indices) > 0:
        print(f"Indices of all-zero rows: {zero_row_indices}")

    # Find non-zero rows
    non_zero_mask = ~(np.all(X == 0, axis=1))

    # Filter adata to only include non-zero rows
    adata = adata[non_zero_mask].copy()

    # Shuffle and assign group labels
    
    adata.obs['shards'] = np.random.choice([f'shard_{i}' for i in range(shards)], size=adata.n_obs)

    adata.write('/maiziezhou_lab2/yunfei/Projects/FM_temp/InterPLM/interplm/ge_shards/cosmx_human_lung_sec8.h5ad')
else:
    print("using cached")
    adata = sc.read_h5ad(p)

model_dir = "/maiziezhou_lab/zihang/SpatialFoundationModel/eval/scGPT/whole-human-pretrain"
model_dir = Path(model_dir)
output_dir = Path("/maiziezhou_lab2/yunfei/Projects/interpTFM/activations_cosmx_lung_cancer")
model_name = "scGPT"
layers = range(12)

process_h5ad(
    adata=adata,
    model_dir=model_dir,
    output_dir=output_dir,
    model_name=model_name,
    layers=layers,
    device='cuda:0',
    shards=shards,
    batch_size=128
)
```
